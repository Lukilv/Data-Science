2. Выберите топ 3 страны по количеству заказчиков, выведите их названия и количество записей.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT  c.Country, COUNT(o.OrderID) as TOTAL_ORDERS FROM orders o
    join customers c on o.CustomerID= c.CustomerID
    GROUP BY Country
    ORDER BY TOTAL_ORDERS DESC
    LIMIT 3;
    ''',
    con,
)
```

3. Выберите перевозчика, который отправил 10-й по времени заказ, выведите его название, и дату отправления.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT s.ShipperName, o.OrderDate
    FROM orders o
    JOIN Shippers s ON o.ShipperID = s.ShipperID
    ORDER BY OrderDate ASC
    LIMIT 1 OFFSET 9
    ''',
    con,
)
```

4. Выберите самый дорогой заказ, выведите список товаров с их ценами.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT o.OrderID, p.ProductName ,p.Price From OrderDetails o
    JOIN Products p on o.ProductID =p.ProductID
    WHERE o.OrderID= (
      SELECT o2.OrderID FROM OrderDetails o2
    join Products p2 ON  o2.productId=p2.ProductID
    GROUP BY o2.OrderID
    ORDER BY SUM(p2.Price * o2.Quantity) DESC
    LIMIT 1

    );

    ''',
    con,
)
```

5. Какой товар больше всего заказывали по количеству единиц товара, выведите его название и количество единиц в каждом из заказов.

6. Выведите топ 5 поставщиков по количеству заказов, выведите их названия, страну, контактное лицо и телефон.

7. Какую категорию товаров заказывали больше всего по стоимости в Бразилии, выведите страну, название категории и сумму.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT c.Country, c1.CategoryName, SUM(p.Price * o.Quantity) as cost FROM Orders o1
    JOIN Customers c ON o1.CustomerID=c.CustomerID
    JOIN OrderDetails o ON o.OrderID=o1.OrderID
    JOIN PRODUCTS p ON p.ProductID=o.ProductID
    JOIN CATEGORIES c1 ON c1.CategoryID=p.CategoryID
    WHERE c.Country ='Brazil'
    AND p.CategoryID=(
          SELECT p2.CategoryID  FROM OrderDetails o2
          join Products p2 ON  o2.productId=p2.ProductID
          GROUP BY p2.CategoryID
          ORDER BY SUM(p2.Price * o2.Quantity) DESC
          LIMIT 1
    );



    ''',
    con,
)
```

8. Какая разница в стоимости между самым дорогим и самым дешевым заказом из США.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT (SELECT SUM(p2.Price * o2.Quantity) FROM OrderDetails o2
    join Products p2 ON  o2.productId=p2.ProductID
    JOIN Orders o1 ON o2.OrderID=o1.OrderID
    JOIN Customers c ON o1.CustomerID=c.CustomerID
    WHERE c.Country ='USA'
    GROUP BY o2.OrderID
    ORDER BY SUM(p2.Price * o2.Quantity) DESC
    LIMIT 1) - (SELECT SUM(p2.Price * o2.Quantity) FROM OrderDetails o2
    join Products p2 ON  o2.productId=p2.ProductID
    JOIN Orders o1 ON o2.OrderID=o1.OrderID
    JOIN Customers c ON o1.CustomerID=c.CustomerID
    WHERE c.Country ='USA'
    GROUP BY o2.OrderID
    ORDER BY SUM(p2.Price * o2.Quantity) ASC
    LIMIT 1) AS difference


    ''',
    con,
)
```

9. Выведите количество заказов у каждого из трех самых молодых сотрудников, а также имя и фамилию во второй колонке.
